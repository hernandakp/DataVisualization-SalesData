```python
'Author: hernanda472@gmail.com'

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Dataset.xlsx')
df
```

```python
df.plot(x="nama_komoditas", kind='bar', figsize=(20, 7))
plt.title('Data penjualan 2019', size=16)
plt.ylabel('Penjualan', size=14)
plt.xlabel('Nama komoditas', size=14)
plt.show()
```

```python
df.plot(x="nama_komoditas", y="november", kind='bar', figsize=(10, 6), color = 'red')
plt.title('Data penjualan di bulan november 2019', size=16)
plt.ylabel('Penjualan', size=14)
plt.xlabel('Nama Komoditas', size=14)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
plt.plot(df['january'], label='january', marker = 'o', color = 'red')
plt.plot(df['february'], label='february', marker = 'o', color = 'blue')
plt.plot(df['march'], label='march', marker = 'o', color = 'green')
plt.title("Data penjualan di bulan januari sampai march 2019", size=20)
plt.xlabel('Nomor nama Komoditas', size=16)
plt.ylabel('Penjualan', size=16)
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

```python
max = df.loc[:, 'january':].max()
plt.figure(figsize=(12, 5))
plt.plot(max, label='penjulan tertinggi', color = 'purple', marker = 'o')
plt.title("Penjualan tertinggi komoditas di tahun 2019", size=20)
plt.xlabel('Bulan', size=16)
plt.ylabel('Penjualan Tertinggi', size=16)
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

```python
max = df.loc[:, 'january':].min()
plt.figure(figsize=(12, 5))
plt.plot(max, label='penjulan terendah', color = 'grey', marker = 'o')
plt.title("Penjualan terendah komoditas di tahun 2019", size=20)
plt.xlabel('Bulan', size=16)
plt.ylabel('Penjualan terendah', size=16)
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

```python
rerata = df.loc[:, 'january':].mean()
plt.figure(figsize=(12, 5))
plt.plot(rerata, label='Rata rata', color = 'orange', marker = 'o')
plt.title("\nRata-rata penjualan komoditas di tahun 2019", size=20)
plt.xlabel('\nBulan', size=16)
plt.ylabel('\nRata-rata penjualan', size=16)
plt.legend(loc='upper left', fontsize=8)
plt.show()
```
